```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'pennines_sin_grad_le_fil.csv'  # Replace with your CSV file's path
data = pd.read_csv(file_path)

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4.5), sharey=True)

# Plot the first scatter plot on the first subplot
ax1.scatter(data['gradient'], data['sinuosity'], alpha=0.7, edgecolor='k')
ax1.set_xlabel('Gradient', fontsize=10)
ax1.set_ylabel('Sinuosity', fontsize=10)
ax1.grid(False)

# Plot the second scatter plot on the second subplot
ax2.scatter(data['m_chi'], data['sinuosity'], alpha=0.7, edgecolor='k')
ax2.set_xlabel('Ksn', fontsize=10)
ax2.grid(False)

# Adjust layout to avoid overlap
plt.tight_layout()

plt.show()
```
